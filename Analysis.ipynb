```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("stats.txt", header= None, names=['score'])
```

```python
df = df [:500]
```

```python
plt.plot(df.score)
plt.ylabel('score')
plt.show()
```

```python
dfa = ((df + df.shift(-1)) / 2)[::50]
```

```python
plt.plot(dfa.score)
plt.ylabel('score')
plt.show()
```
